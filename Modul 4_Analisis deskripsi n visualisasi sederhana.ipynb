# **Analisis Visual**

### Perbandingan antar kategori


```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')


plt.figure(figsize=(10, 6))
sns.boxplot(x='Type', y='Price', data=df)
plt.title('Distribution of Property Prices by Type')
plt.xlabel('Property Type')
plt.ylabel('Price')
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(x='Method', y='Price', data=df)
plt.title('Distribution of Property Prices by Method')
plt.xlabel('Sale Method')
plt.ylabel('Price')
plt.show()


top_regions = df['Regionname'].value_counts().nlargest(10).index
df_top_regions = df[df['Regionname'].isin(top_regions)]

plt.figure(figsize=(12, 8))
sns.boxplot(x='Regionname', y='Price', data=df_top_regions)
plt.title('Distribution of Property Prices by Top 10 Regions')
plt.xlabel('Region Name')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Distribusi data


```python
plt.figure(figsize=(18, 5))

# Histogram for Price
plt.subplot(1, 3, 1)
sns.histplot(df['Price'], bins=50, kde=True)
plt.title('Distribution of Property Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Histogram for Distance
plt.subplot(1, 3, 2)
sns.histplot(df['Distance'], bins=50, kde=True)
plt.title('Distribution of Distance from City')
plt.xlabel('Distance (km)')
plt.ylabel('Frequency')

# Histogram for Landsize
plt.subplot(1, 3, 3)
sns.histplot(df['Landsize'], bins=50, kde=True)
plt.title('Distribution of Landsize')
plt.xlabel('Landsize (sqm)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Korelasi antar kolom


```python
numerical_cols = df.select_dtypes(include=np.number)

correlation_matrix = numerical_cols.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Columns')
plt.show()
```

## Hubungan antar variabel



```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Price vs. Rooms
sns.scatterplot(ax=axes[0, 0], x='Rooms', y='Price', data=df)
axes[0, 0].set_title('Price vs. Number of Rooms')
axes[0, 0].set_xlabel('Number of Rooms')
axes[0, 0].set_ylabel('Price')

# Price vs. Distance
sns.scatterplot(ax=axes[0, 1], x='Distance', y='Price', data=df)
axes[0, 1].set_title('Price vs. Distance from City')
axes[0, 1].set_xlabel('Distance from City (km)')
axes[0, 1].set_ylabel('Price')

# Price vs. Landsize
sns.scatterplot(ax=axes[1, 0], x='Landsize', y='Price', data=df)
axes[1, 0].set_title('Price vs. Landsize')
axes[1, 0].set_xlabel('Landsize (sqm)')
axes[1, 0].set_ylabel('Price')

# Price vs. BuildingArea
sns.scatterplot(ax=axes[1, 1], x='BuildingArea', y='Price', data=df)
axes[1, 1].set_title('Price vs. Building Area')
axes[1, 1].set_xlabel('Building Area (sqm)')
axes[1, 1].set_ylabel('Price')

plt.tight_layout()
plt.show()
```
